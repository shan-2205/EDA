#### Merge data from each month into one CSV

```python
#search on stackoverflow to merge data


path = "F:/EDA_projects/Sales_Analysis/SalesAnalysis/Sales_Data"
files = [file for file in os.listdir(path)]

#blank dataframe
all_months_data = pd.DataFrame()

for file in files:
    current_df = pd.read_csv(path+"/"+file)
    all_months_data = pd.concat([all_months_data, current_df])
    
all_months_data.shape
    
```

#  Add hour and minute column

```python

all_data['Hour'] = pd.to_datetime(all_data['Order Date']).dt.hour
all_data['Minute'] = pd.to_datetime(all_data['Order Date']).dt.minute
```

```python
all_data
```

```python
def hour(x):
    c=x.split(' ')[1]
    return c.split(':')[0]
```

```python
all_data['hour']=all_data['Order Date'].apply(hour)
```

```python
all_data
```

```python
def minute(x):
    c=x.split(' ')[1]
    return c.split(':')[1]
```

```python
all_data['minute']=all_data['Order Date'].apply(minute)
```

```python
all_data
```

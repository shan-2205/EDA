###  2(b) : (II) Popular Platforms to Host / Run LLM Models : Hugging Face , Groq !
                But why these platforms exist ?
                
                Bcz many LLMs are extremely large ‚Äî sometimes tens or hundreds of GBs,and in some cases even TBs in size..
                ( like LLaMA 70B , DeepSeek LLM etc..) 
                To run such big models, you need :

                a) Powerful GPUs
                b) High processing power
                c) Huge storage
                
                & Most users don‚Äôt have that hardware on their laptop or PC.
        
                Thats why Platforms like Hugging Face and Groq host these large models in the cloud, which means:

                a) You don‚Äôt need to download the model
                b) You don‚Äôt need expensive hardware

                You can simply send a request (API call) and u can use these models...

## Lecture 2(c) : Evolution: ML ‚Üí DL ‚Üí LLMs(GenAI)

```python
'''

üåü Lecture : Evolution : ML ‚Üí DL ‚Üí LLMs(GenAI)

    -> Machine Learning
        Machine or Models or Algo wll actually learn patterns from the data..
        In ML , u have to manually give data to train your Algo !
        ie , you have to tell ML Algo them what features matter.‚Äù

        Input ‚Üí model ‚Üí output
        
        eg , Your phone uses ML to detect whether a message or call is spam or not spam.
        Bcz ML model inside your phone has been trained on specific data like Sender info , Message length , Frequency of 
        msg/call



    -> Deep Learning
        Here Models or Algo learn patterns automatically from raw data, using neural networks..
        U dont have to tell manually to your Algo what features matter..
        DL Algos Works well with big data

        eg , Your phone‚Äôs face unlock actually uses DL algo to unlock phone as DL algo inside your phone lock has been trained on 
        huge amt of Facial data , thats why it can unlock your phone instantly and accurately. !



    -> Large Language Models (LLMs)
        LLMs are advanced DL models trained on internet-scale text so they can understand and generate human language.‚Äù
        They can actually understand and respond like humans , eg they can Write essays, emails, stories , Generate code , 
        compose music..
        Examples:

        GPT
        LLaMA
        Gemini
        Claude


    ‚≠ê One-Line Summary : 
        ML learns patterns manually , DL learns automatically, LLMs generate human-like content..


'''
```

```r
"

duplicated(x, incomparables = FALSE, ...)

## Default S3 method:
duplicated(x, incomparables = FALSE,
           fromLast = FALSE, nmax = NA, ...)

## S3 method for class 'array'
duplicated(x, incomparables = FALSE, MARGIN = 1,
           fromLast = FALSE, ...)






1.. duplicated(x, incomparables = FALSE, ...)
    duplicated() in R is used to identify duplicate elements 
    in a vector, data frame or array

    x : The vector, array, or data frame to check for duplicates..

    incomparables : A vector of values that should be considered as 
                    unique and not compared for duplication..
                            OR
                    A vector of values that should not be considered 
                    when determining duplicates.
                    Default is FALSE..




2.. Default S3 method:
    duplicated(x, incomparables = FALSE,
           fromLast = FALSE, nmax = NA, ...)


    This is the default method of the duplicated function, suitable 
    for most objects like vectors, data frames, etc.
    In simple terms in R, S3 stands for "Simple, Structured, and Sensible." 
    It's a basic system for object-oriented programming
                OR
    In R, S3 methods are a way to implement object-oriented programming.



    x : The vector, array, or data frame to check for duplicates..

    incomparables : A vector of values that should be considered 
                    as unique and not compared for duplication..
                            OR
                    A vector specifying values that should not be considered 
                    when determining duplicates.
                    Default is FALSE..

    fromLast : A logical value indicating whether to identify duplicates 
                from the last occurrence. 
               Default is FALSE..

    nmax : An integer specifying the maximum number of duplicates to return. 
           Default is NA, meaning all duplicates are returned..




3.. S3 method for class 'array':
    duplicated(x, incomparables = FALSE, MARGIN = 1,
           fromLast = FALSE, ...)

    This method is specifically for arrays.
    It identifies duplicates along a specific margin (dimension) of the array.


    MARGIN: An integer indicating which margin to check for duplicates. 
            Default is 1 (rows).
            columns (if MARGIN = 2)


"
```

```r
args(aggregate)

## it doesn't give much better arguments list so lets use help(aggregate)
```

```r
help(aggregate)

## check definition of aggregate() using help()
```

```r
"

aggregate(x, ...)

## Default S3 method :
aggregate(x , ...)

## S3 method for class 'data.frame'
aggregate(x , by , FUN , ..., simplify = TRUE , drop = TRUE)

## S3 method for class 'formula'
aggregate(formula , data , FUN , ..., subset, na.action = na.omit)

## S3 method for class 'ts'
aggregate(x , nfrequency = 1 , FUN = sum , ndeltat = 1 , 
            ts.eps = getOption("ts.eps"), ...)


"
```

```r
"

The aggregate function in R has multiple outputs because 
it is a versatile function with different methods depending 
on the type of data you're working with. 
Let's break down each method briefly:


1.. Default S3 method :
    aggregate(x, ...)

    This is the generic form of the aggregate function that can work 
    with various types of data objects.
    It aggregates values of x according to the grouping 
    specified in the ... arguments.

    In simple terms in R, S3 stands for "Simple, Structured, and Sensible." 
    It's a basic system for object-oriented programming
                OR
    In R, S3 methods are a way to implement object-oriented programming.



2.. S3 method for class 'data.frame'
    aggregate(x, by, FUN, ..., simplify = TRUE, drop = TRUE)

    This method is specifically designed to work with data frames.
    It aggregates values of columns in a data frame according 
    to the grouping specified by the "by" argument.
    
    x = col on which I want to apply aggregation function ..
    by = grouping variables
    FUN = aggregation function.




3.. S3 method for class 'formula':
    aggregate(formula , data , FUN , ..., subset , na.action = na.omit)

    This method is used when you are working with dataframes & 
    you provide a formula specifying the variables to aggregate 
    and the grouping structure.

    formula =  It is a symbolic description of the form y ~ x1 + x2 + ..., 
               where y is the variable to be summarized, 
                and x1, x2, etc., are the variables to group by..
    data = data frame containing the variables.
    FUN = aggregation function.




4.. S3 method for class 'ts':
    aggregate(x , nfrequency = 1 , FUN = sum , ndeltat = 1 , 
    ts.eps = getOption("ts.eps") , ...)

    This method is for time series objects (ts).
    It aggregates values in a time series object over specific frequencies 
    (nfrequency) using a specified function (FUN).

    x = col on which I want to apply aggregation function ..
    nfrequency = frequency for aggregation
    FUN = aggregation function.


"
```

```r
### u can use both 1st , 2nd & 3rd one to perform group-by , 
### so lets use 3rd one !
```

#### Syntax : aggregate(formula , data , FUN , ...)

    formula : It is a symbolic description of the form y ~ x1 + x2 + ..., 
                where y is the variable to be summarized, and x1, x2, etc., are the variables to group by..

    data : data frame containing the variables..

    FUN : aggregation function..

```r
aggregate(Total_Guests ~ arrival_date_month , data , sum)

## aggregate(data$Total_Guests, by = list(data$arrival_date_month), 
    FUN = sum , simplify = TRUE, drop = TRUE)

## data %>% group_by(arrival_date_month) %>% 
## summarise(Total_guests = sum(Total_Guests))
```

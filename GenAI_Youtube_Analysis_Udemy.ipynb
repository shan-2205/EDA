        ie Understand Viewer Emotions !

````python
'''

Sentiment analysis is all about finding sentiment scores(polarity) which range between (-1,+1)
more it is close to -1 , ie more it is negative ( This is the worst movie ever)
more close to 0 means ,, it is neutral sentence ( This movie is Okay)
more it is +1 , more it is +ve ( This case study is amazing )


'''
````

````python
'''
‚ùå Traditional

    NLTK
    VADER
    Loops



'''
````

````python
### we are going to use LLMs to perform Sentiment Analysis  , Because LLMs understand human language..
````

````python
'''

LLMs are trained on:

    Millions of reviews
    Social media posts
    Product feedback
    News articles
    
    So the model has seen similar sentences thousands of times..
    Thats why it can perform Sentiment Analysis easily..



'''
````

````python
def llm_sentiment(text):
    prompt = f"""
    You are a sentiment analysis system.

    STRICT RULES:
    - Return ONLY raw JSON
    - Do NOT add text, markdown, or explanation
    - Do NOT wrap in ```json
    - If unsure, still return neutral

    Text: "{text}"

    JSON format:
    {{
      "sentiment": "positive | negative | neutral",
      "score": number between -1 and 1
    }}
"""

    response = llm.invoke(prompt)

    return response.content
````

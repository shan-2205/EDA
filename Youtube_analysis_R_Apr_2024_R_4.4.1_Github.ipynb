```r
"

DocumentTermMatrix (documents: 1 , terms : 739):
This is a Document-Term Matrix (DTM), where each row represents a document and each column represents a term (word) from the text.
documents: 1 means there is only one document in this matrix.
terms: 739 means there are 739 unique terms (words) in the document.

Non-/sparse entries: 739/0:
Non-sparse entries: There are 739 entries that are non-zero, 
                    meaning each term appears at least once in this document.
Sparse entries: There are no sparse entries (0), meaning 
                there are no terms that have a frequency of zero in this particular document..

Sparsity : 0%:
Sparsity refers to the percentage of zeroes in the matrix. 
Since this matrix has 0% sparsity, all terms have non-zero values..

Maximal term length: 42:
This indicates the length of the longest term (word) in the matrix, which in this case is 42 characters.

Weighting : term frequency (tf):
Weighting describes the method used to fill the matrix. 
Here, it‚Äôs term frequency (tf), meaning each cell in the matrix 
contains the frequency (count) of each term in the document.




Summary : DocumentTermMatrix represents a single document with 739 unique terms, 
          where all terms appear at least once (0% sparsity),
          the longest term is 42 characters, and the matrix uses term frequency as its weighting scheme..


"
```

```r
"

emojis are just Unicode characters , meaning each emoji has a specific Unicode code point. 
For example, the emoji for a smiling face üòä is represented by the code point U+1F60A..




U+ Prefix:

The U+ prefix indicates that this is a Unicode code point. 
Unicode code points are typically represented in the format U+XXXX, where XXXX is a hexadecimal number.


1F60A:
This is the actual hexadecimal number that represents the emoji.

"
```

```r
help(gsub)

## Pattern Matching and Replacement

### It stands for "Global substitution(gsub)
```

```r
"

gsub(pattern , replacement , x , ignore.case = FALSE , perl = FALSE,
     fixed = FALSE , useBytes = FALSE)


pattern : This is the pattern you want to search for within the string.

replacement : This is the string that will replace the matched occurrences of the pattern..

x : This is the input string (or vector of strings) in which replacements are to be made..

ignore.case : A logical value. If TRUE, it ignores case when matching patterns.

fixed : A logical value. If TRUE, pattern is treated as a string literal instead of a regular expression.

perl : A logical value. If TRUE, uses Perl-compatible regular expressions.

"
```

```r
args(gsub)
```

```r
gsub("[[:punct:]]", "", full_df$title[2])

### this code removes all the punc from the above text or it replaces punc with empty string !
### ie where-ever this pattern got matches , replace it with ""
```

```r
### Now I will say remove everything except punc.. (so that I will get all the puncs)
```

```r
gsub("[^[:punct:]]", "", full_df$title[2])

### It removes all non-punctuation characters from the first row of the 'title' column.
```

```r
"


[^ ... ]: This is a character class, indicated by the square brackets [ ]. 
          It matches any character that is not listed within the brackets.
          ie , ^ - negation of the pattern

[:punct:]: Inside the character class, [:punct:] is a special character class that represents all punctuation characters. 
           Punctuation characters typically include symbols like periods, commas, exclamation marks, 
            question marks, and various other characters that are not letters or digits.

         So, when you use [^[:punct:]] as a regular expression pattern, 
         ie , it matches any character that is not a punctuation character. 
         For example, it would match characters , letters, digits, spaces, and any non-punctuation characters..

        ie in a short , it replaces everything with empty string except punc


"
```

```r
nchar(gsub("[^[:punct:]]", "", full_df$title[2])) ## ie total 3 punc 
```

```r
count_punctuation = function(x)
{
  return(nchar(gsub("[^[:punct:]]", "", x)))
}
```

```r
count_punctuation(full_df$title[3])
```

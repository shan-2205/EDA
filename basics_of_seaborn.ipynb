## what is Dataframe ??
    Its all about data in the form of Rows & colums  ,
    ie It is exactly equal to csv data (excel)
    If you're familiar with the Table in Excel then you already have a grasp of what a Pandas dataframe is. 
    It's a table of data. 
    It has rows and columns
    Its 2-D data structure
    Its like a Table or a matrix
    If we see internally , implementation of pandas dataframe is all about ' Sequence of series '
    ie ,if series is all about list or array
    then dataframe is all bout lists of list..


    From Python - point of view ,for those of u who
     are from python background ,we have data-structure called Dictionary..
    It is just a extension of dictionary (data-structure of python) 
    e.g->> dict1={'first_name':'Shan' ,
                  'last_name':'Singh'
                  }
    in dataframe ,key becomes feature & value becomes rows..


    Dataframe are size-mutable,  heterogeneous tabular data

     Pandas DataFrame will be created by loading the 
     data sets from various storage: 
     storage can be SQL Database, .CSV file, and Excel file. 
     Pandas DataFrame can also be created internally 
     from the lists, dictionary, and from a list of dictionary etc.




     Series & dataframe are 2 different classes in pandas.

## attributes of pandas dataframe..
    1..values -->> Return a Numpy representation of the DataFrame.
    2..shape -->> Return a tuple representing the dimensions of the DataFrame.
    3..dtypes-->> returns data-type of the features of a dataframe
    4..columns-->> returns column-name of the DataFrame.

    5..index -->> returns The index (row labels) of the DataFrame.
    6..iloc -->> integer-location based indexing for selection by position.
    7..loc -->> Access a group of rows and columns by label(s) or a boolean array.
    8..at -->> Access a single value for a row/column label pair.
    9..iat-->> Access a single value for a row/column pair by integer position.

    10..axes -->> Return a list representing the axes of the DataFrame.
    11..ndim -->> Return an int representing the number of axes / array dimensions.
    12..size -->> Return an int representing the number of elements in this object.

    13..empty-->> Indicator whether DataFrame is empty.
    14..style-->> Returns a Styler object.
    15..attrs-->> Dictionary of global attributes of this dataset.
    16..flags-->> Get the properties associated with this pandas object.

    17..T -->> returns tranpose of a dataframe..

### customizing width & bins in histogram

```python
sns.histplot(df['rate'])

##  plt.hist(df['rate'])
```

```python
The towers or bars of a histogram are called bins 

The height of each bin shows how many values from 
                   that data fall into that range. 
    
In seaborn , it chooses best value of bin_size automatically
using bunch of heuristics..

In matplotlib ,the default value of this bin parameter is 10..

As the bin size is changed so then the bin width 
                would be changed accordingly as :
Width of each bin is = 
(max value of data  min value of data) / total number of bins     
    
    
### bins = also called "intervals"  or "buckets".
```

```python
sns.histplot(df['rate'],bins=20)

## ie ,I need total 20 bars or 20 towers
## plt.hist(df['rate'] ,bins=20)
```

### How to set title in seaborn

```python
'''
In pandas, date & time values are stored using a special data type called : datetime


Common datetime formats you‚Äôll see
    datetime64[ns]
    datetime64[ns, tz]
    datetime64[ns, UTC]
    <M8[ns]
    
    Don‚Äôt panic ‚Äî they look different, but conceptually they are the same family..




datetime64[ns]
    datetime64 ‚Üí date + time data
    ns ‚Üí nanoseconds (very high precision)
    64 means : It takes 64 bytes/8 bits to store date-time values
    üëâ datetime64[ns] = date-time stored with nanosecond precision



datetime64[ns, UTC]
    What is UTC?

    UTC = Coordinated Universal Time
    It‚Äôs the global standard time
    No country bias (used in aviation, servers, finance)

    So:
    üëâ datetime64[ns, UTC] = date-time stored in UTC time zone


datetime64[ns, TZ]
    tz = time zone
    Example :

    Asia/Kolkata
    US/Eastern

    So:
    üëâ datetime64[ns, TZ] : date-time with time zone information


What is <M8[ns]?
    This is just a low-level NumPy representation.

    M ‚Üí datetime type
    8 ‚Üí 8 bytes (64 bits)
    ns ‚Üí nanoseconds
    < ‚Üí little-endian (how bytes are stored in memory)



'''
```

```python
#### Converting the flight Dep_Time into proper time i.e. mid_night, morning, afternoon and evening.

def flight_dep_time(x):
    '''
    This function takes the flight Departure time 
    and convert into appropriate format.
    
    '''
    
    if (x>4) and (x<=8):
        return "Early Morning"
    
    elif (x>8) and (x<=12):
        return "Morning"
    
    elif (x>12) and (x<=16):
        return "Noon"
    
    elif (x>16) and (x<=20):
        return "Evening"
    
    elif (x>20) and (x<=24):
        return "Night"
    
    else:
        return "late night"
```

```python
'''
we have some rows in "Duration" col which have values as :

19h

55 mins

12 hrs 24 mins


very first we have to preprocess it so that we will extract duration hour & duration mins 


'''
```

```python
def preprocess_duration(x):
    if 'h' not in x:
        x = '0h' + ' ' + x
    elif 'm' not in x:
        x = x + ' ' +'0m'
        
    return x
```

```python
data['Duration'] = data['Duration'].apply(preprocess_duration)
```

```python
data['Duration']
```

```python
'''
    Now after pre-processing duration feature , still my ml_model is not able to understand duration 
    bcz it is string data so any how we have to convert it into numerical(integer of float) values

'''
```

```python
'''
1Ô∏è‚É£ Dataset Size ‚Äî Why does it matter?
One-Hot Encoding creates new columns

Example:
Airline ‚Üí IndiGo, Air India, Vistara


One-Hot becomes:
Airline_IndiGo  Airline_AirIndia  Airline_Vistara


Small dataset ‚úÖ

    Few rows
    Few categories
    Model can handle extra columns easily

üëâ No problem.


Large dataset ‚ùå

    Millions of rows
    Many categories
    One-Hot creates hundreds or thousands of columns



This causes:

    High memory usage
    Slower training
    Overfitting

    üëâ That‚Äôs why One-Hot is risky for large datasets.


'''
```

```python
'''
2Ô∏è‚É£ Cardinality ‚Äî Why number of unique values matters?
    Low cardinality (safe)

    Example:

    Total_Stops ‚Üí non-stop, 1 stop, 2 stops
    Only 3 values ‚Üí One-Hot is fine.


    High cardinality (danger)
    Example:
    Route ‚Üí 400+ unique routes


    One-Hot would create:
    400+ new columns üò±


    This:

    Bloats the dataset
    Adds noise

    üëâ That‚Äôs why for high-cardinality features we prefer:

    Label Encoding
    Frequency Encoding
    Target Encoding



'''
```

```python
'''
1Ô∏è‚É£ What is an outlier ?
    An outlier is a price that is very different from most other prices.

    Example:

    Most flights cost ‚Üí ‚Çπ4,000 ‚Äì ‚Çπ12,000
    Suddenly you see ‚Üí ‚Çπ75,000

    üëâ That ‚Çπ75,000 is an outlier.


2Ô∏è‚É£ Why do outliers exist in flight prices?
    
    Outliers can happen because of:

    Business class / premium tickets
    Last-minute bookings
    Data entry errors

    So:

    Not all outliers are wrong
    But they can confuse ML models
    
    
3Ô∏è‚É£ Why do we handle outliers?   
    [ 5000,  5600 , 6400 , 6600 , 6800 , 7100 , ]
    Average price ‚âà ‚Çπ6,000
    
    [ 5000,  5600 , 6400 , 6600 , 6800 , 7100 , 65000 ]
    Average price ‚âà ‚Çπ14,000
    üëâ Model now overestimates normal flight prices
    



'''
```

### Hyperparameter Tuning or Hyperparameter Optimization
    1.Choose following method for hyperparameter tuning
        a.RandomizedSearchCV --> Fast way to Hypertune model

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
### initialise your estimator
reg_rf = RandomForestRegressor()
```

```python
np.linspace(start =100 , stop=1200 , num=6)
```

```python
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start =100 , stop=1200 , num=6)]

# Number of features to consider at every split
max_features = ["auto", "sqrt"]

# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(start =5 , stop=30 , num=4)]

# Minimum number of samples required to split a node
min_samples_split = [5,10,15,100]
```

```python
# Create the random grid or hyper-parameter space

random_grid = {
    'n_estimators' : n_estimators , 
    'max_features' : max_features , 
    'max_depth' : max_depth , 
    'min_samples_split' : min_samples_split
}
```

```python
random_grid
```

```python
# Define searching

# Random search of parameters, using 3 fold cross validation
# search across 576 different combinations


rf_random = RandomizedSearchCV(estimator=reg_rf , param_distributions=random_grid , cv=3 , n_jobs=-1 , verbose=2)
```

# What is LLM ?

```python
'''

üîç Let‚Äôs break down the term LLM :

üü¢ 1Ô∏è‚É£ Large

‚ÄúLarge‚Äù means two things:

a) Huge training data
LLMs are trained on massive amounts of text, like:

books
articles
websites
documentation
code

This helps the model learn grammar, facts, patterns, and writing styles.

b) Huge number of parameters

Parameters are internal numbers the model uses to make predictions.
More parameters ‚Üí model can learn more complex patterns.
Modern LLMs have billions of parameters.




üü¢ 2Ô∏è‚É£ Language

This means the model works with human language.

It can:

read text
understand meaning (to some extent)
generate sentences
translate
summarize
answer in conversation form

It doesn‚Äôt truly ‚Äúunderstand‚Äù like humans ‚Äî it works by recognizing patterns in language.





üü¢ 3Ô∏è‚É£ Model

A model is a mathematical system trained to make predictions.

In LLMs:
üëâ The prediction is usually ‚ÄúWhat is the next word?‚Äù

Example:
‚ÄúData analysis is used to find meaningful ‚Ä¶‚Äù

Model predicts: ‚Äúinsights‚Äù

By predicting one word at a time very fast, it creates full sentences and answers.


In short : An LLM is a very large AI model trained on massive text data that predicts the next word 
            to generate human-like language.


'''
```

```python
'''
‚öôÔ∏è Basic Concepts of LLMs (Must Know) :

üîπ 1Ô∏è‚É£ Tokens

LLMs don‚Äôt read full sentences at once.
They break text into small pieces called tokens.

A token can be:

a word
part of a word
punctuation

Example:
‚ÄúData analytics is powerful‚Äù

Might become:
[‚ÄúData‚Äù, ‚Äú analytics‚Äù, ‚Äú is‚Äù, ‚Äú powerful‚Äù]




üîπ 2Ô∏è‚É£ Context Window

This is the memory limit of the model in one conversation.
It means:
How much text the model can look at at one time while answering.

Bigger context window ‚Üí model remembers more of your earlier conversation.




üîπ 3Ô∏è‚É£ Training

LLMs are trained using a method called:

üëâ Next Word Prediction

The model reads a sentence with a missing word and tries to guess it.
Over billions of examples, it learns language patterns.




üîπ 4Ô∏è‚É£ Prompt

A prompt is simply the instruction you give to the LLM.

Better prompt ‚Üí better output.

Example:
‚ùå ‚ÄúExplain this data‚Äù
‚úÖ ‚ÄúExplain this sales data in 5 bullet points for a business manager‚Äù





üîπ 5Ô∏è‚É£ Temperature (Creativity Control)

Temperature controls how creative vs predictable the output is.

Low temperature ‚Üí more factual, stable answers
High temperature ‚Üí more creative, varied answers




üîπ 6Ô∏è‚É£ Hallucination

LLMs sometimes give confident but wrong answers.
This is called hallucination.

That‚Äôs why:
üìå Humans must verify important outputs.

'''
```

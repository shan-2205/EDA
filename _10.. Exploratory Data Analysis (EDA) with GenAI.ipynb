# 1.. What is EDA ?

```python
'''


EDA = understand dataset to make conclusions

üßë‚Äçüíª In traditional EDA:

    We manually create charts (bar chart, histogram, boxplot, etc.)
    We look at numbers and graphs
    We think and write insights ourselves

It takes time and experience to find good insights.


ü§ñ GenAI-Powered EDA 
    AI suggests what plots we can use
    AI quickly finds patterns and trends
    AI gives summary insights in simple language


üéØ Most Important Point
EDA is not just about making graphs.

Graphs are tools ‚Äî
üëâ The real goal is to find insights that help in decision-making




Types of EDA :

Univariate : Looking at one variable at a time
Bivariate : Looking at two variable at a time
Multi-variate : Looking at three or more variables together




Your columns in sales data : 
city, sales, discount, quantity, category, profit

1Ô∏è‚É£ Univariate EDA
    Sales ‚Üí What is the average sale amount?
    Profit ‚Üí Are most orders profitable or loss-making?
    Category ‚Üí Which product category appears most?
    
2Ô∏è‚É£ Bivariate EDA
    Sales vs Profit ‚Üí Do higher sales always mean higher profit?
    Discount vs Profit ‚Üí Does more discount reduce profit?
    City vs Sales ‚Üí Which cities generate higher sales?
    
3Ô∏è‚É£ Multivariate EDA
    City + Category + Sales ‚Üí Which category sells most in each city?
    City + Discount + Sales ‚Üí Do some cities respond better to discounts?


'''
```

# 2.. Univariate analysis using AI !

# 4.. Correlation Insights using AI

```python
'''

üìå What is Correlation ?

Correlation means how two things move together.

If one increases and the other also increases ‚Üí Positive correlation
If one increases and the other decreases ‚Üí Negative correlation
If there is no clear pattern ‚Üí No correlation


üí∞ Sales & Profit
If sales increase and profit also increases, they have a positive correlation.
üëâ More revenue is leading to more profit.

üéÅ Discount & Profit
If discount increases and profit decreases, they have a negative correlation.
üëâ Giving too much discount is hurting profit.


üì¶ Quantity & Sales
If quantity sold increases and sales increase, that‚Äôs a positive correlation.
üëâ Selling more items leads to higher sales value.




üß† Simple Business Meaning

Correlation helps us understand:

What increases profit
What reduces profit
Where we should give discounts
What drives higher sales

'''
```

```python
corr = sales_data[["Sales", "Profit", "Discount", "Quantity"]].corr()
print(corr)
```

```python
## Ask GenAI to explain correlation
```

```python
prompt = f"""
You are a Data Analyst.

Explain correlation results in simple English.

Correlation matrix:
{corr}

Rules:
- 5 bullet points only
- Mention business meaning
- Beginner-friendly
"""
```

```python
response = llm.invoke(prompt)
print(response.text)
```

# 5.. Auto EDA summary prompt !

```python
'''

We give basic dataset info to GenAI and it gives EDA summary..
What is that info :

    total rows
    total columns
    total missing vales
    Descriptive stats of data ( mean , median , std dev , quantile values)

'''
```

```python
eda_info = {
    "rows": sales_data.shape[0],
    "columns": list(sales_data.columns),
    "missing_values": sales_data.isnull().sum().to_dict(),
    "basic_stats": sales_data[["Sales", "Profit", "Discount", "Quantity"]].describe().to_dict()
}
```

```python


prompt = f"""
You are a Data Analyst.

Generate an EDA summary report.

Dataset info:
{eda_info}

Report must include:
1) Dataset overview
2) Top 5 patterns
3) Top 5 issues to check
4) 5 recommended charts

Keep answer short and simple.
"""

```

```python
response = llm.invoke(prompt)
```

```python
print(response.content)
```

```python
response = llm.invoke(prompt)
print(response.text)
```
